### 1. What is Matplotlib? 

üî∏ [Matplotlib](https://matplotlib.org/) is a **plotting library** for the Python programming language

üî∏ It provides an **objected-oriented API** for creating beautiful plots

<a name="2"></a>
### 2. Why Matplotlib? 

üî∏ Built on **NumPy** array (and Python)

üî∏ Integrates directly with Pandas

üî∏ Can create **basic** to **advances** plots

üî∏ Simple to use interface

```python
# import 'matplotlib' package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot()
```

```python
plt.plot()
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [5, 10, 15, 20]
```

```python
plt.plot(x, y)
plt.show()
```

**Note**

> There is **two** ways of creating plots using **matplotlib**; The **pyplot** API is generally less-flexible than the **objected-oriented** API

**Note**

>In general, try to use the **object-oriented** interface over the **pyplot** interface

```python
# First method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
# Second method
fig = plt.figure() # creates a figure
ax = fig.add_axes([1, 1, 1, 1])
plt.show()
```

```python
# Third method (recommended)
fig, ax = plt.subplots()
ax.plot(x, y) # add some data
plt.show()
```

<a name="line-plot"></a>
#### Line Plot

```python
# Plot the data (default is line plot)
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(x, y)
plt.show()
```

<a name="scatter-plot"></a>
#### Scatter Plot

```python
# Use the same data to create 'scatter' plot
fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(x, y)
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots(figsize=(7, 5))
ax.scatter(x, np.cos(x))
plt.show()
```

<a name="bar-plot"></a>
#### Bar Plot

```python
# Create a plot from dictionary
items_inventory = {
    "t-shirt": 10,
    "black shoes": 20,
    "skirt": 12
}
```

```python
items_inventory
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
ax.bar(x=items_inventory.keys(),
       height=items_inventory.values())
ax.set(title="Inventory Status",
       xlabel="item name",
       ylabel="inventory")
fig.savefig("Jupyter_Images/inventory_status.png")
plt.show()
```

```python
# Create 'barh' plot
fig, ax = plt.subplots(figsize=(7, 5))
ax.barh(y=items_inventory.keys(),
        width=items_inventory.values())
ax.set(title="Items inventory",
       xlabel="inventory count",
       ylabel="item name")
plt.show()
```

<a name="histogram-plot"></a>
#### Histogram Plot

```python
# Create 'histogram' plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

<a name="subplots"></a>
#### Subplots

üî∏ There are **two** ways that we can create **multiple** plots.

##### Option one (Recommended)

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))

plt.show()
```

```python
# Plot to each different axis
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
ax1.plot(np.arange(0, 10), np.arange(0, 10)**2)
ax2.scatter(x, x**2)
ax3.bar(items_inventory.keys(), items_inventory.values())
ax4.hist(x)
plt.show()
```

##### Option Two

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
ax1 = ax[0, 0]
ax2 = ax[0, 1]
ax3 = ax[1, 0]
ax4 = ax[1, 1]

ax1.plot(np.arange(10), np.arange(10)**2)
ax2.scatter(x, x**2)
ax3.bar(items_inventory.keys(), items_inventory.values())
ax4.hist(x)
plt.show()
```

**Note** : `ax` variable is type of **NumPy ndarray** it's not a Python's list

```python
import pandas as pd
```

üëÄ In the following cell we've used **RegEx** to remove punctuation from our **price** columns

üîó [Learn RegEx here](https://regexone.com)

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True).str[:-2]
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
```

```python
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods=10)
```

```python
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
```

```python
car_sales
```

```python
# Let's plot the total sale
car_sales.plot(x="Sale Date",
               y="Total Sales")

plt.show()
```

```python
# Let's have 'scatter' plot
car_sales.plot(x="Odometer (KM)",
               y="Price",
               kind="scatter")
plt.show()
```

```python
# Let's have a 'bar' graph
car_sales
```

```python
car_sales.plot(x="Make",
               y="Price",
               kind="bar")

plt.show()
```

```python
car_sales.plot(kind="bar")
plt.show()
```

```python
# Let's have 'histogram' plot of Odometer (km)
car_sales["Odometer (KM)"].plot(kind="hist",
                                bins=10)
plt.show()
```

### 6. Working with Heart Disease Dataset

```python
# import 'heart_disease.csv'
heart_disease = pd.read_csv("data/heart-disease.csv")
```

```python
heart_disease.head()
```

```python
# Create a 'histogram' plot
heart_disease["age"].plot(kind="hist",
                          bins=50)

plt.show()
```

<a name="7"></a>
### 7. Matplotlib OO Method with Pandas

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50
```

```python
# Using 'pyplot' method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target")

plt.show()
```

```python
# Using OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(7, 5))
over_50.plot(kind="scatter",
                   x="age",
                   y="chol",
                   c="target",
                   ax=ax)

# ax.set_xlim([45, 100])
plt.show()
```

<a name="8"></a>
### 8. Customizing Our Plots Professionally

```python
over_50.head(3)
```

```python
# Using OO method from scratch
fig, ax = plt.subplots(figsize=(7, 5))

# Plotting the data
scatter_plot = ax.scatter(x=over_50["age"],
                          y=over_50["chol"],
                          c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease Risk Based on Cholesterol Level",
       xlabel="age",
       ylabel="cholesterol")

# Add a legend
ax.legend(*scatter_plot.legend_elements(), title="target")

# Add a horizontal line to the plot
ax.axhline(over_50["chol"].mean(),
           linestyle="--",
           linewidth=1)

plt.show()
```

<a name="9"></a>
### 9. Subplots


- **thalach** ‚Üí The person's **maximum heart rate** achieved

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 12),
                               sharex=True)

# Add data to ax1:
scatter_plot1 = ax1.scatter(x=over_50["age"],
                            y=over_50["chol"],
                            c=over_50["target"])
ax1.set(title="Heart Disease Risk Based on Cholesterol Level",
        ylabel="cholesterol")

ax1.legend(*scatter_plot1.legend_elements(),
           title="target")

ax1.axhline(over_50["chol"].mean(),
            linestyle="--",
            linewidth=1)

# Add data to ax2:
scatter_plot2 = ax2.scatter(x=over_50["age"],
                            y=over_50["thalach"],
                            c=over_50["target"])

ax2.set(title="Heart Disease Risk Based on Maximum Heart Rate",
        xlabel="age",
        ylabel="maximum heart rate")

ax2.legend(*scatter_plot2.legend_elements(),
           title="target")

ax2.axhline(over_50["thalach"].mean(),
            linestyle="--",
            linewidth=1)

# Add a title to the whole figure
fig.suptitle("Heart Disease Analysis",
            fontsize=12,
             fontweight="bold")

plt.show()
fig.savefig("Jupyter_Images/heart_disease_risk_based_on_chol_&_thalach.png")
```

<a name="10"></a>
### 10. Changing The Default Style

```python
# Available style for matplotlib
plt.style.available
```

```python
plt.style.use("seaborn-v0_8")
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 12),
                              sharex=True)
# Let's plot on ax1
scatter_plot1 = ax1.scatter(x=over_50["age"],
                            y=over_50["chol"],
                            c=over_50["target"],
                            cmap="tab10")

ax1.set(title="Heart Disease Risk Based on Cholesterol Level",
        ylabel="cholesterol")

ax1.legend(*scatter_plot1.legend_elements(),
           title="target")

ax1.axhline(over_50["chol"].mean(),
            linestyle="--",
            linewidth=1,
            color="red")
# Change ax1 x axis limitation
ax1.set_xlim([50, 80])


# Let's plot on ax2
scatter_plot2 = ax2.scatter(x=over_50["age"],
                            y=over_50["thalach"],
                            c=over_50["target"],
                            cmap="Set3")

ax2.set(title="Heart Disease Risk Based on Maximum Heart Rate",
        xlabel="age",
        ylabel="maximum heart rate")

ax2.legend(*scatter_plot2.legend_elements(),
           title="target")

ax2.axhline(over_50["thalach"].mean(),
            linestyle="--",
            linewidth=1,
            color="red")
# Change ax1 y axis limitation
ax2.set_ylim([60, 200])

# Add a title to the whole figure
fig.suptitle("Heart Disease Analysis",
            fontsize=12,
             fontweight="bold")

plt.show()
```
